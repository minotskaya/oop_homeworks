```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
b_row_1 = []
for x in [1., 15.]:
    b_row_1.append(f(x))
```

```python
a1 = np.array([[1, 1],
              [1, 15]])
b1 = np.array(b_row_1).T
```

```python
w1 = np.linalg.solve(a1, b1)
```

```python
def pol_1(w, x):
    return np.array(w[0]+w[1]*x)
```
